```python
def take_input():
    n, c = input().split()
    n = int(n)
    c = int(c)
    interval=[]
    for j in range(n):
        l, r = input().split()
        l = int(l)
        r = int(r)
        interval.append([l,r])
    new_interval = total_new_intervals(interval,c)
    total_intervals = new_interval + n
    return total_intervals
```

```python
def total_new_intervals(interval, c):
    h_map = [0]*((10**5)+1)
    h_map
    for i in interval:
        for j in range((i[0]+1), i[1]):
            print(j)
            h_map[j] = h_map[j]+1
    h_map.sort()
    h_map.reverse()
    new_interval=sum(h_map[:c])
    return new_interval
```
