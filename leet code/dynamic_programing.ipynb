```python
list_of_number =[i for i in range(2,1000)]
# list_of_number
```

```python
for i in list_of_number:
    print(f'******************************************************************************')
    print(f'first i = {i}')
    num_to_remove=i+i
    print(f'first num_to_remove = {num_to_remove}')
    while num_to_remove<1000:
        if num_to_remove in list_of_number:
            list_of_number.remove(num_to_remove)
        num_to_remove=num_to_remove+i
        print(f'inside while num_to_remove = {num_to_remove}')
```
